```python

import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import requests
from pandas.io.json import json_normalize
import json

import requests

from bs4 import BeautifulSoup

from geopy.geocoders import Nominatim

import folium
import matplotlib.cm as cm
import matplotlib.colors as colors

from sklearn.cluster import KMeans
```

### Part 1: Web Scraping to get a cleaned dataframe of Toronoto Neighbourhood from the wikipedia page

```python
#Using Beautiful Soup to open the wikipedia page
webdata=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(webdata, 'html.parser')
```

```python
soup
```

```python
#Creating empty lists for storing the scraped web data
postalCodeList = []
boroughList = []
neighborhoodList = []

for row in soup.find('table').find_all('tr'):
    cells = row.find_all('td')
    if(len(cells) > 0):
        postalCodeList.append(cells[0].text)
        boroughList.append(cells[1].text)
        neighborhoodList.append(cells[2].text.rstrip('\n'))
```

```python
#Creating a dataframe from the web data
toronto_city = [('PostalCode', postalCodeList),
                      ('Borough', boroughList),
                      ('Neighborhood', neighborhoodList)]
toronto_df = pd.DataFrame.from_dict(dict(toronto_city))
toronto_df.head()
```

```python
#Removing postal codes which are not assigned to any boroughs
toronto_df_new = toronto_df[toronto_df.Borough != 'Not assigned'].reset_index(drop=True)
toronto_df_new.head()
```

```python
#Grouping Neighbourhood that belong to the same postal code and borough
toronto_df_grouped = toronto_df_new.groupby(['PostalCode','Borough'], as_index=False).agg(lambda x: ','.join(x))
toronto_df_grouped.head()
```

```python
#Replacing an unassigned neighbourhood with its borough name
na_neigh_rows = toronto_df_grouped.Neighborhood == 'Not assigned'
toronto_df_grouped.loc[na_neigh_rows, 'Neighborhood'] = toronto_df_grouped.loc[na_neigh_rows, 'Borough']
toronto_df_grouped[na_neigh_rows]
```

```python
#Displaying the cleaned dataframe
toronto_df_clean=toronto_df_grouped
toronto_df_clean.shape
```

### Part 3: Exploring and Clustering Neighbourhoods in Toronto

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(toronto_df_coords['Borough'].unique()),
        toronto_df_coords.shape[0]
    )
)
```

```python
#Getting the latitude and longitude of Toronto 
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="tl-toronto-neigh")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geographical coordinates of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
#Using Folium to create a map of Toronnto with boroughs as markers
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

for lat, long, post, borough, neigh in zip(toronto_df_coords['Latitude'], toronto_df_coords['Longitude'], toronto_df_coords['PostalCode'], toronto_df_coords['Borough'], toronto_df_coords['Neighborhood']):
    label = "{} ({}): {}".format(borough, post, neigh)
    popup = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, long],
        radius=5,
        popup=popup,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)
    
map_toronto
```

```python
#Selecting Boroughs with Toronto in their names
toronto_boroughs = ['East Toronto', 'Central Toronto', 'Downtown Toronto', 'West Toronto']
toronto_df_latest = toronto_df_coords[toronto_df_coords['Borough'].isin(toronto_boroughs)].reset_index(drop=True)
print(toronto_df_latest.shape)
toronto_df_latest.head()
```

```python
ap_toronto = folium.Map(location=[latitude, longitude], zoom_start=12)

for lat, long, post, borough, neigh in zip(toronto_df_latest['Latitude'], toronto_df_latest['Longitude'], toronto_df_latest['PostalCode'], toronto_df_latest['Borough'], toronto_df_latest['Neighborhood']):
    label = "{} ({}): {}".format(borough, post, neigh)
    popup = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, long],
        radius=5,
        popup=popup,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)
    
map_toronto
```

```python
#Getting places inside and near Toronto central
radius = 500
LIMIT = 100

venues = []

for lat, long, post, borough, neighborhood in zip(toronto_df_latest['Latitude'], toronto_df_latest['Longitude'], toronto_df_latest['PostalCode'], toronto_df_latest['Borough'], toronto_df_latest['Neighborhood']):
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    for venue in results:
        venues.append((
            post, 
            borough,
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
venues_df = pd.DataFrame(venues)
venues_df.columns = ['PostalCode', 'Borough', 'Neighborhood', 'BoroughLatitude', 'BoroughLongitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']
print(venues_df.shape)
venues_df.head()
```

```python
venues_df.groupby('Borough').count()
```

```python
print('There are {} uniques categories.'.format(len(venues_df['VenueCategory'].unique())))
```

```python
#Analyzing venues in each area

# one hot encoding
toronto_df_latest_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# adding postal, borough and neighborhood column back to dataframe
toronto_df_latest_onehot['PostalCode'] = venues_df['PostalCode'] 
toronto_df_latest_onehot['Borough'] = venues_df['Borough'] 
toronto_df_latest_onehot['Neighborhoods'] = venues_df['Neighborhood'] 

# moving postal, borough and neighborhood column to the first column
fixed_columns = list(toronto_df_latest_onehot.columns[-3:]) + list(toronto_df_latest_onehot.columns[:-3])
toronto_df_latest_onehot = toronto_df_latest_onehot[fixed_columns]

print(toronto_df_latest_onehot.shape)
toronto_df_latest_onehot.head()
```

```python
#Getting frequency of occurences of each category in an area
toronto_df_latest_venues_freq = toronto_df_latest_onehot.groupby(['PostalCode', 'Borough', 'Neighborhoods']).mean().reset_index()
print(toronto_df_latest_venues_freq.shape)
toronto_df_latest_venues_freq.head()
```

```python
#Getting the 10 most oocurence venue type in each area
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
areaColumns = ['PostalCode', 'Borough', 'Neighborhoods']
freqColumns = []
for ind in np.arange(num_top_venues):
    try:
        freqColumns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        freqColumns.append('{}th Most Common Venue'.format(ind+1))
columns = areaColumns+freqColumns

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['PostalCode'] = toronto_df_latest_venues_freq['PostalCode']
neighborhoods_venues_sorted['Borough'] = toronto_df_latest_venues_freq['Borough']
neighborhoods_venues_sorted['Neighborhoods'] = toronto_df_latest_venues_freq['Neighborhoods']

for ind in np.arange(toronto_df_latest_venues_freq.shape[0]):
    row_categories = toronto_df_latest_venues_freq.iloc[ind, :].iloc[3:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    neighborhoods_venues_sorted.iloc[ind, 3:] = row_categories_sorted.index.values[0:num_top_venues]

neighborhoods_venues_sorted.sort_values(freqColumns, inplace=True)
neighborhoods_venues_sorted
```

```python
#Using K-means to get 5 clusters
kclusters = 3

toronto_df_latest_venues_freq_clustering = toronto_df_latest_venues_freq.drop(['PostalCode', 'Borough', 'Neighborhoods'], 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_df_latest_venues_freq_clustering)

toronto_df_latest_clustered_df = toronto_df_latest
toronto_df_latest_clustered_df['Cluster'] = kmeans.labels_

toronto_df_latest_clustered_df = toronto_df_latest_clustered_df.join(neighborhoods_venues_sorted.drop(['Borough', 'Neighborhoods'], 1).set_index('PostalCode'), on='PostalCode')
toronto_df_latest_clustered_df.sort_values(['Cluster'] + freqColumns, inplace=True)
toronto_df_latest_clustered_df
```

```python
# creating map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# setting color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# adding markers to the map
markers_colors = []
for lat, lon, post, bor, poi, cluster in zip(toronto_df_latest_clustered_df['Latitude'], toronto_df_latest_clustered_df['Longitude'], toronto_df_latest_clustered_df['PostalCode'], toronto_df_latest_clustered_df['Borough'], toronto_df_latest_clustered_df['Neighborhood'], toronto_df_latest_clustered_df['Cluster']):
    label = folium.Popup('{} ({}): {} - Cluster {}'.format(bor, post, poi, cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
